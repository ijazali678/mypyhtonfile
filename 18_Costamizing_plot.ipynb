```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import data
flower= sns.load_dataset("iris")
kashti= sns.load_dataset("titanic")
tip = sns.load_dataset("tips")
```

```python
kashti.head()
```

```python
sns.boxplot(x="survived", y= "age", data= kashti)
plt.show
```

```python
#means show means
sns.boxplot(x="survived", y= "age", showmeans=True,data= kashti,)
plt.show
```

```python
sns.boxplot(x="survived", y= "age", showmeans=True,
meanprogs=["marker": "+",
            "markersize":"12",
            "markercolor":"red"], data= kashti)
plt.show
```

```python
#lables for publish graph
# sns.boxplot(x="survived", y= "age", showmeans=True,
# data= kashti)
grid=sns.FacetGrid(data=kashti,col="age", hue="age", palette="tab20c",
                     col_wrap=4, height=1.5)
grid.refline(y=0, linestyle=":")
grid.map(plt.plot, "sibsp", "fare", marker="o")
# Adjust the tick positions and labels
grid.set(xticks=np.arange(5), yticks=[-3, 3],
         xlim=(-.5, 4.5), ylim=(-3.5, 3.5))

# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=1)
plt.xlabel("Survived peson")
plt.ylabel("age of person")
plt.title("box plot of titanic ", size=20, weight ="bold")

plt.show
```
