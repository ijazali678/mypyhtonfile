```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
flower= sns.load_dataset("iris")
```

```python
sns.barplot(x="species", y= "sepal_width", data= flower)
plt.show
```

# now for grouping you mast need different datasets

```python
kashti= sns.load_dataset("titanic")
```

```python
kashti
```

```python
sns.barplot(x="sex", y= "alone", hue= "who", data= kashti)
plt.show
```

## Horizantal plot

```python
#you have to just change the x and yaxis of catogries
sns.barplot(x="alone", y= "sex", hue= "who", data= kashti)
plt.show
```

# Now how to set ordering the group 

```python
sns.barplot(x="sex", y= "alone", hue= "who", data= kashti, 
order=["female", "male"], saturation=1)
plt.show
```

# for colouring colour pelet

```python
sns.barplot(x="sex", y= "alone", hue= "who", data= kashti, 
order=["female", "male"], color= "red")
plt.show
```

```python
sns.barplot(x="sex", y= "alone", hue= "who", data= kashti, 
order=["female", "male"], color= "red", ci=None) #ci = None is used for eror bor
plt.show
```

```python
# already develp pllet seaborn palette
#some example are "tab10", "pastel", "muted", "bright", "colorblind", "dark"
# "Set2", "husl", "hls", "Set2", "paired", "rocket", "mako"

sns.barplot(x="sex", y= "alone", hue= "who", data= kashti, 
order=["female", "male"], color= "red", ci=None,
palette="mako") #ci = None is used for eror bor
plt.show
```

```python
# estimator we use as a numerical value n saturation
from numpy import mean
sns.barplot(x="sex", y= "fare", hue= "sex", data= kashti, 
order=["female", "male"], color= "red", estimator=mean,
palette="mako", saturation=1) #ci = None is used for eror bor
plt.show
```
