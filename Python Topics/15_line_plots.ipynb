## Line and bar plots

the first thing is to import libraries 
 - numpay
 - seaborn
 - pandas
 - matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Now load the dataset

```python
flower= sns.load_dataset("iris")
```

```python
flower
```

now draw a line plot

```python
sns.lineplot(x="sepal_length", y= "sepal_width", data= flower)
plt.show
```

# Now how to write a tital

```python
sns.lineplot(x="sepal_length", y= "sepal_width", data= flower)
plt.title("plot of phool")
plt.show
```

## Adding limits

```python
sns.lineplot(x="sepal_length", y= "sepal_width", data= flower)
plt.title("plot of phool")
plt.xlim(2)
plt.ylim(1)
plt.show
```

# Now how to set style
- dark
- white 
- ticks
- whitegride

```python
sns.set_style(style=None, rc = None)
#this was assignment and done
#before changing of style you mast set_style as None
sns.lineplot(x="sepal_length", y= "sepal_width", data= flower)
plt.title("plot of phool")
sns.set_style("dark")
plt.show
```

# Now size and figure

```python
plt.figure(figsize= (10,3))
sns.lineplot(x="sepal_length", y= "sepal_width", data= flower)
plt.title("plot of phool")
sns.set_style("dark")
```
